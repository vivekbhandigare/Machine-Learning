```python
import numpy as np
import matplotlib.pyplot as plt
```

# Z = 2*X*X - 3*Y*Y + 5*e

```python
x = np.arange(-1, 1, step=0.01)
y = np.arange(-1, 1, step=0.01)
```

```python
len(x), len(y)
```

```python
X,Y = np.meshgrid(x,y)
```

```python
X.shape, Y.shape
```

```python
c = np.ones((200,200))
```

```python
e = np.random.rand(200,200)*0.1
```

```python
Z = 2*X*X - 3*Y*Y + 5*e
```

```python
Z.shape
```

```python
import vis
```

```python
vis.plot3d(X,Y,Z)
```

```python
X.shape, Y.shape, Z.shape
```

```python
input_xy = np.c_[X.reshape(-1), Y.reshape(-1)]
```

```python
output_z = Z.reshape(-1)
```

```python
input_xy.shape, output_z.shape
```

# model

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
```

```python
model.add(Dense(40, input_dim=2, activation="relu"))
model.add(Dense(20, activation="relu"))
model.add(Dense(20, activation="relu"))

model.add(Dense(1))
```

```python
model.summary()
```

```python
model.compile(loss = "mean_squared_error", optimizer = "sgd")
```

```python
output = model.fit(input_xy, output_z, epochs=20, validation_split=0.1)
```

# predict

```python
z_pred = model.predict(input_xy).reshape(200,200)
```

```python
vis.plot3d(X,Y, z_pred)
```
